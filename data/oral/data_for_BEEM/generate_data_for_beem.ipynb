```python
interaction = res_df['value'][ntaxa:].to_numpy()
interaction = interaction.reshape(ntaxa, ntaxa)   # row index -- source taxa, column index -- target taxa
```

```python
metadata = pd.read_csv("metadata_{}.txt".format(ntaxa), sep='\t')
```

```python
metadata
```

```python
patient_ids = list(set(metadata['subjectID']))
patient_ids = [int(pid) for pid in patient_ids]
```

```python
num_patients = len(patient_ids)
num_patients
```

```python
num_days_per_patient = [sum(metadata['subjectID'] == pid) for pid in patient_ids]
```

```python
sum(num_days_per_patient)
```

```python
def alr_to_p(alr_vec, reference_idx):
    """alr_vec: (d, )
        p: (d, )
    """
    tmp = np.exp(alr_vec)
    tmp[...,reference_idx] = 1
    return tmp/tmp.sum(axis=-1, keepdims=True)
    
    
def p_to_alr(p_vec, reference_idx):
    pseudo = 1e-6
    p_vec[p_vec==0] = pseudo
    return np.log(p_vec / p_vec[..., reference_idx:reference_idx+1])
    

def p_to_tilde_p(p_vec, reference_idx):
    pseudo = 1e-6
    p_vec[p_vec==0] = pseudo
    return p_vec / p_vec[..., reference_idx:reference_idx+1]
```

```python
def forward(pvec, biomass, growth_rate, interaction_matrix, reference_idx):
    #print(pvec.shape)
    tilde_p = p_to_tilde_p(pvec, reference_idx)
    interaction_term = np.matmul(interaction_matrix, tilde_p.T)
    interaction_term = interaction_term.T   # (bsz, d)
    alr = np.log(tilde_p)
    alr_next = alr + growth_rate + biomass[:,None] * interaction_term # (d, )
    #print("alr shape", alr.shape)
    #print("alr_next shape", alr_next.shape)
    alr_next[..., reference_idx] = 0
    pvec_next = alr_to_p(alr_next, reference_idx)
    
    return pvec_next
    
```

```python
def n_step_prediction(pvecs, biomass_arr, time_idx, max_n_steps, growth_rate, interaction_matrix, reference_idx):
    """
    pvecs: (total_days, d)
    time_idx: (total_days, )
    max_n_steps: integer
    return: 
    prediction_list, a list of length max_n_steps, each list is of shape (total_time-k_step, d) where k_step in range(max_n_steps)
    mse, a dict, {time_idx:  mse_array}  / TODO: or return a mask? 
    """
    total_time = int(time_idx[-1] - time_idx[0] + 1)
    ndays, d = pvecs.shape
    prediction_arr = np.empty((max_n_steps, ndays, d))
    
    previous_vecs = pvecs
    for k_step in range(max_n_steps):
        pvecs_next = forward(previous_vecs, biomass_arr, growth_rate, interaction_matrix, reference_idx)
        assert pvecs_next.shape == pvecs.shape, "{}}, {}".format(pvecs.shape, pvecs_next.shape)
        prediction_arr[k_step] = pvecs_next
        previous_vecs = pvecs
        
    return prediction_arr
        
```

```python
# load counts data
counts_df = pd.read_csv('counts_{}.txt'.format(ntaxa), sep='\t')
```

```python
counts = counts_df.iloc[0:,1:].to_numpy()
```

```python
counts_per_patient = []
percentage_per_patient = []
previous_idx = 0
for num_days_p in num_days_per_patient:
    c = counts[:, previous_idx : previous_idx+num_days_p].T
    p = c / c.sum(axis=-1, keepdims=True)
    counts_per_patient.append(c)
    percentage_per_patient.append(p)
    previous_idx = previous_idx + num_days_p
```

```python
# load biomass
biomass_df = pd.read_csv("biomass_{}.csv".format(ntaxa), dtype=np.float)
```

```python
biomass = biomass_df['biomass'].to_numpy()
biomass.shape
```

```python
biomass_per_patient = []
previous_idx = 0
for num_days_p in num_days_per_patient:
    biomass_per_patient.append(biomass[previous_idx : previous_idx+num_days_p])
    previous_idx = previous_idx + num_days_p
```

```python
# compute time index
time_index_per_patient = []
previous_idx = 0
for num_days_p in num_days_per_patient:
    time_index_per_patient.append(metadata['measurementID'][previous_idx:previous_idx+num_days_p].to_numpy())
    previous_idx = previous_idx + num_days_p
```

```python
time_index_per_patient[0].shape
```

```python
max_n_steps = 5
prediction_list = []
for pid in patient_ids:
    # (max_n_steps, ndays, ntaxa)
    prediction_t = n_step_prediction(pvecs=percentage_per_patient[pid], biomass_arr=biomass_per_patient[pid], 
                                    time_idx=time_index_per_patient[pid], max_n_steps=max_n_steps, 
                                    growth_rate=growth_rate, interaction_matrix=interaction,
                                    reference_idx=reference_idx)
    prediction_list.append(prediction_t)
```

```python
# collect 1 step

def get_rsq(pid, max_n_steps):
    ytrue_list = [np.empty((0, ntaxa)) for _ in range(max_n_steps)]
    yhat_list = [np.empty((0, ntaxa)) for _ in range(max_n_steps)]
    prediction_t = prediction_list[pid]
    for k_step in range(max_n_steps):
        #print("k_step  = ", k_step+1)
        for k_step_ in range(max_n_steps):
            #print("actual gap = ", k_step_ + 1)
            has_k_step = np.where(
                time_index_per_patient[pid][k_step_+1:] - time_index_per_patient[pid][:-k_step_-1] == k_step+1)[0]
            #print(has_k_step)
            if len(has_k_step) > 0:
                pred_idx = has_k_step + k_step_ + 1
                ytrue = percentage_per_patient[pid][pred_idx]
                yhat = prediction_t[k_step][has_k_step]
                ytrue = np.concatenate([ytrue_list[k_step], ytrue], axis=0)
                yhat = np.concatenate([yhat_list[k_step], yhat], axis=0)
                ytrue_list[k_step] = ytrue
                yhat_list[k_step] = yhat
    
    rsq_list = {}
    for k_step in range(max_n_steps):
        if len(ytrue_list[k_step]) == 0:
            print(" No {} step for patient {}".format(k_step+1, pid))
        else:
            mse = np.sum((ytrue_list[k_step] - yhat_list[k_step])**2)
            yi_mean = np.mean(ytrue_list[0])
            yi_var = np.sum((ytrue_list[0] - yi_mean)**2)
            rsq = 1 - mse/(yi_var)
            rsq_list[k_step] = rsq
    return rsq_list
```

```python
rsqs_for_all_patients = [[] for _ in range(max_n_steps)]
for pid in patient_ids:
    rsq_t = get_rsq(pid, max_n_steps)
    for k_step in range(max_n_steps):
        if k_step in rsq_t:
            if not np.isnan(rsq_t[k_step]) and not np.isinf(rsq_t[k_step]):
                rsqs_for_all_patients[k_step].append(rsq_t[k_step])
```

```python
rsqs_for_all_patients = [np.array(rsq_kstep) for rsq_kstep in rsqs_for_all_patients]
```

```python
mean_rsqs = [np.mean(tmp) for tmp in rsqs_for_all_patients]
```

```python
plt.plot(range(1, max_n_steps+1), mean_rsqs, '-o')
plt.title('{} taxa'.format(ntaxa))
plt.ylabel("R square")
plt.xlabel('prediction step')
plt.savefig("{}taxa_kstep.jpg".format(ntaxa))
```
